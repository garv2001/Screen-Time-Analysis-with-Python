```python
import seaborn as sns
```

```python
sns.set(style="darkgrid")
bar=sns.barplot(x="Date",y="Usage",data=screen,hue="App")
bar.set_xticklabels(bar.get_xticklabels(),rotation=85)
plt.title("Usage")
plt.show()

```

```python

bar_1=sns.barplot(x="Date",y="Times opened",data=screen,hue="App",palette="dark")
bar_1.set_xticklabels(bar_1.get_xticklabels(),rotation=85)
plt.title("Times Opened")
plt.show()
```

```python
#Notifications
bar_2=sns.barplot(x="Date",y="Notifications",data=screen,hue="App")
bar_2.set_xticklabels(bar_2.get_xticklabels(),rotation=85)
plt.title("Notifications")
plt.show()
```

```python
bar_3=sns.scatterplot(x="Notifications",y="Usage",data=screen,hue="App")
#bar_3.set_xticklabels(bar_3.get_xticklabels())
plt.title("Relationship between Number of Notifications and Usage")
plt.show()
```

```python
bar_3=sns.regplot(x="Notifications",y="Usage",data=screen)
#bar_3.set_xticklabels(bar_3.get_xticklabels())
plt.title("Relationship between Number of Notifications and Usage")
plt.show()
```

```python
#Notification and Usage
bar_3=sns.lmplot(x="Notifications",y="Usage",data=screen,hue="App")
#bar_3.set_xticklabels(bar_3.get_xticklabels())
plt.title("Relationship between Number of Notifications and Usage for Different Apps")
plt.show()
```
